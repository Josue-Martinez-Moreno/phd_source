```python
import sys
from netCDF4 import Dataset
import os
import cmocean as cm
from trackeddy.tracking import *
from trackeddy.datastruct import *
from trackeddy.geometryfunc import *
from trackeddy.init import *
from trackeddy.physics import *
from trackeddy.plotfunc import *
from numpy import *
from pylab import *
import cmocean as cm
from calendar import monthrange
%matplotlib inline
```

```python
years=range(1993,2017)
index=0
eke_mean=zeros(365*len(years))
eke_noeq=zeros(365*len(years))
eke_south=zeros(365*len(years))
eke_north=zeros(365*len(years))
for yy in years:
    for mm in range(1,13):
        datashapetime=monthrange(int(yy), mm)[1]
        print(yy,mm)
        for dd in range(1,datashapetime):
            inputfiles='/g/data/ua8/CMEMS_SeaLevel/v3-0/'+str(yy)+'/'
            
            try:
                ncfile=Dataset(inputfiles+'dt_global_allsat_phy_l4_'+str(yy)+\
                           '%02d'% mm+'%02d'% dd+'_20170110.nc')
                ssha=squeeze(ncfile.variables['sla'][:])
                lon=ncfile.variables['longitude'][:]
                lat=ncfile.variables['latitude'][:]
                mask=ma.getmask(ssha[:,:])
                u_g,v_g=geovelfield(ssha,lon,lat,mask)
                eke=KE(u_g,v_g)
                eke_mean[index]=np.nanmean(eke)
                eke[350:370,:]=nan
                eke_noeq[index]=np.nanmean(eke)
                eke_south[index]=np.nanmean(eke[0:350,:])
                eke_north[index]=np.nanmean(eke[370::,:])
            except:
                print('File doesnt exist'+inputfiles+'dt_global_allsat_phy_l4_'+str(yy)+\
                           '%02d'% mm+'%02d'% dd+'_20170110.nc')
            index=index+1       
```

```python
eke_mean[eke_mean==0]=nan
eke_noeq[eke_noeq==0]=nan
eke_south[eke_south==0]=nan
eke_north[eke_north==0]=nan
```

```python
figure()
pcolormesh(lon,lat,eke[:,:],vmin=0,vmax=0.2)
```

```python


base = datetime.datetime(1993, 1, 1, 0, 0)
date_list = [base + datetime.timedelta(days=x) for x in range(0, len(eke_mean))]

fig, ax = plt.subplots(figsize=(15,3))
ax.plot(date_list,eke_mean,'-g',label='$KE$')
ax.plot(date_list,eke_noeq,'-k',label='$KE_e$')
ax.plot(date_list,eke_south,'-r',label='$KE_south$')
ax.plot(date_list,eke_north,'-b',label='$KE_north$')
ax.grid(True)
ax.set_xlim(date_list[0],date_list[-1])

monthsFmt = DateFormatter("%b/%y")
ax.xaxis.set_major_formatter(monthsFmt)
```
