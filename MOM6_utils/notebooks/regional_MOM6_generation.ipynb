```python
from numpy import *
from pandas import *
from xarray import *
from pylab import *
import cmocean as cm
%matplotlib inline
```

```python
depth=open_dataset('../../data.input/GEBCO/GEBCO_2014_2D.nc')
```

```python
new_domain=depth.where(depth.elevation<0).elevation.sel(lat=slice(-80, -60),lon=slice(155,185))
new_domain.plot.pcolormesh(cmap=cm.cm.balance)
```

```python
interpix=depth.sel(lon=arange(155,180,0.05),method='nearest')
```

```python
interpiy=interpix.sel(lat=arange(-80,-55,0.05),method='nearest')
```

```python
interpiy.where(interpiy.elevation<0).elevation.plot.pcolormesh(cmap=cm.cm.deep_r)
```
