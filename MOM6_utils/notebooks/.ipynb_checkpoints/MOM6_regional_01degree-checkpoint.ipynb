```python
from scipy import interpolate
from numpy import *
from pandas import *
from xarray import *
from pylab import *

worldt=open_dataset('WOA05_pottemp_salt.nc')

worldt.load() 

#worldt['TIME']=date_range('2000-01-15', periods=12, freq='M')
time=[15, 44, 73.5, 104, 134.5, 165, 195.5, 226.5, 257, 287.5, 318.5, 349]
worldt['TIME']=time
worldt2=worldt.sel(LON=arange(0,360,0.1),method='nearest').sel(LAT=linspace(-90,90,2700),method='nearest')
#worldt2=worldt2.fillna(-1.e+34)
worldt2['LON']=arange(0,360,0.1)
worldt2['LAT']=linspace(-90,90,2700)

lon=arange(0,360,0.1)
lat=linspace(-90,90,2700)
newfield=zeros([12,33,len(lat),len(lon)])
for tt in range(0,12):
    print(tt)
    for zz in range(0,33):
        Lon,Lat=meshgrid(lon,lat)
        LonI,LatI=meshgrid(worldt.LON.data,worldt.LAT.data)
        array = np.ma.masked_invalid(worldt.PTEMP[tt,zz,:,:].data.ravel())
        #get only the valid values
        x1 = LatI.ravel()[~array.mask]
        y1 = LonI.ravel()[~array.mask]
        newarr = array[~array.mask]
        newfield[tt,zz,:,:] = interpolate.griddata((x1,y1), newarr, (Lat,Lon), method='nearest')

        
ptemp=newfield        
for tt in range(0,12):
    print(tt)
    for zz in range(0,33):
        Lon,Lat=meshgrid(lon,lat)
        LonI,LatI=meshgrid(worldt.LON.data,worldt.LAT.data)
        array = np.ma.masked_invalid(worldt.SALT[tt,zz,:,:].data.ravel())
        #get only the valid values
        x1 = LatI.ravel()[~array.mask]
        y1 = LonI.ravel()[~array.mask]
        newarr = array[~array.mask]
        newfield[tt,zz,:,:] = interpolate.griddata((x1,y1), newarr, (Lat,Lon), method='nearest')
        #f=interpolate.interp2d(worldt.LON.data.ravel(),worldt.LAT.data.ravel(),worldt.PTEMP[tt,zz,:,:].data.ravel(),kind='linear')
        #newfield[tt,zz,:,:]=f(lat,lon)
salt=newfield
worldt2['PTEMP']=(('TIME', 'DEPTH', 'LAT', 'LON'),ptemp)
worldt2['SALT']=(('TIME', 'DEPTH', 'LAT', 'LON'),salt)

worldt2.to_netcdf('nonan_test_global.nc')
```
